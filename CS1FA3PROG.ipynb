```python
#fa3program2

score = int(input("Enter student's Exam score: "))

if score > 100:
  print("Invalid input please try again")

elif score < 0:
  print("Invalid input please try again")

elif score >= 75:
  print("Passed")

elif score <= 74:
  print("Failed")

else:
  print("Invalid input please try again")
```
